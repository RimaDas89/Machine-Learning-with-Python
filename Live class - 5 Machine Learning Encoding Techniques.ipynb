```python
# FIlling all the null value with median

ds['age'] = ds['age'].fillna((ds['age'].median()))
ds
```

```python
# Another way to fill na with numpy

import numpy as np

ds = ds.replace(np.NaN, ds['salary'].mean())
ds
```

```python
#Applying label encoder in the Nation column

le = LabelEncoder()
ds['Nation'] = le.fit_transform(ds['Nation'])
ds['Nation']
```
