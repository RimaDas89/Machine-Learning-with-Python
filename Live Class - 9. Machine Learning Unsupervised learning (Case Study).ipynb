# Unsupervised Learning - Customer Segmentation

## K-mean clustering Case study

ContextThis notebook Is created for the learning purpose of the customer segmentation concepts, also known as market basket analysis. I will demonstrate this by using unsupervised ML technique (KMean clustering Algorithm) in the simplest form.

Content you are owing a supermarket mall and through membership cards,  you have some basic data about your customers Like customer ID, age, gender, annual income and spending score. Spending score is something you assign to a customer based on your defined parameterslike customer behaviour and purchasing data.

Problem Statement You own the mall and want to understand the customers like who can be easily converge.[Target customers] so that the sense can be given to marketing team and plan strategy accordingly.

```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.cluster import KMeans

import warnings
warnings.filterwarnings('ignore')
```

```python
dataset = pd.read_csv("Mall_customers.csv")
dataset.head(10)
```

```python
#dataset
x=dataset.iloc[:,3:]
x.tail(8)
```

```python
#elbow method

# kmean++ : selects initial cluster centers for k-mean clustering in a smart way to speed up convergence.

from sklearn.cluster import KMeans

wcss=[]
for i in range(1,11):
    km =KMeans(n_clusters =i, init ="k-means++")
    km.fit(x)
    wcss.append(km.inertia_)
    
plt.plot(range(1,11), wcss, marker = 'x')
plt.title('The Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel("WCSS")
plt.show()
```

```python
#fitting K-means to the dataset

kmeans =KMeans(n_clusters=5, init = "k-means++")
y_kmeans =kmeans.fit_predict(x)
print(y_kmeans)
```

```python
kmeans.labels_
```

```python
#plotting all the clusters and their centroids

#x-axis: annual income
#y-axis: spending score
#s =market size

plt.figure(figsize=(8,8))
plt.scatter(x[y_kmeans==0]['Annual Income (k$)'], x[y_kmeans==0]['Spending Score (1-100)'], s=100, c='green', label='cluster 1')
plt.scatter(x[y_kmeans==1]['Annual Income (k$)'], x[y_kmeans==1]['Spending Score (1-100)'], s=100, c='blue', label='cluster 2')
plt.scatter(x[y_kmeans==2]['Annual Income (k$)'], x[y_kmeans==2]['Spending Score (1-100)'], s=100, c='cyan', label='cluster 3')
plt.scatter(x[y_kmeans==3]['Annual Income (k$)'], x[y_kmeans==3]['Spending Score (1-100)'], s=100, c='violet', label='cluster 4')
plt.scatter(x[y_kmeans==4]['Annual Income (k$)'], x[y_kmeans==4]['Spending Score (1-100)'], s=100, c='yellow', label='cluster 5')

#plot the centroids
#cluster_centers_ndarray od shape(n_clusters, n_features) : Coordinates of cluster centers.

plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=100, c='red', label='centroids')
plt.title("Customer Group")
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```
